### 8/15 Pandas

```python
import pandas as pd
df = pd.DataFrame(columns = ["Column1","Column2"])
df
```

```python
df.columns = ['first','second']
df
```

```python
new_data = {
    "first": [1, 2, 'foo'],
    "second": [3, 4, 'bar'],
}

df2 = pd.DataFrame(new_data)
df2
```

```python
df3 = {
    "first": [5, 6, 'hello'],
    "second": [7, 8, 'World']
}
df3 = pd.DataFrame(df3)
df3
```

```python
df_combined = pd.concat([df2, df3])
df_combined
```

```python
df_combined.reset_index(drop = True, inplace = True)
df_combined
```

```python
column_data = {
    "third": ["a", "b", "c"]
}
```

```python
df4 = pd.DataFrame(column_data)
df_combined = pd.concat([df_combined, df4], axis=1)
df_combined
```

```python
df_sales = {
    "Date": ["Jan 1", "Jan 2"],
    "Amount": [500, 600]
}
df_sales = pd.DataFrame(df_sales)

df_sales2 = {
    "Date": ["Jan 3", "Jan 4"],
    "Amount": [400, 700]
}
df_sales2 = pd.DataFrame(df_sales2)

df_new = pd.concat([df_sales, df_sales2])
df_new.reset_index(drop=True, inplace=True)
df_new

new_column_data = {
    "Location": ["Store A", "Store B"]
}

df_column_data = pd.DataFrame(new_column_data)

df_new = pd.concat([df_new, df_column_data], axis=1)
df_new
```
